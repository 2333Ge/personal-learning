```python
import openai

# 设置查询

# 发送查询请求

# 问题，变量的定义
def getResponse (input):
    prompt = "在末尾我将给你由一对---符号包裹的字符串，将其处理成我想要的形式，要求如下：\
    1. 该字符串是类似JSON的数据结构，可能有语法错误，将其处理为正确的JSON数据结构，暂且将其定称为变量right_json\
    2. 将right_json 转换成typescript 类型声明，这个结果我们称之为result_ts\
    输出形式为：\
    处理后的数据结构：\
    right_json\
    转换成typescript类型\
    result_ts\
    ---{input}---"
    response = openai.Completion.create(
    engine="davinci",
    prompt=prompt,
    max_tokens=60,
    n=1,
    stop=None,
    temperature=0.5,
    )
    return response.choices[0].text.strip()
print(
getResponse('{"name": "John Doe","age": 30,"email": "johndoe@example.com","address": {"street": "123 Main St","city": "Anytown","state": "CA","zip": "12345"},"phoneNumbers": [{"type": "home","number": "555-555-1234"},{"type": "work","number": "555-555-5678"}]}')
)
```
